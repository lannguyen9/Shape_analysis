# Harris Corner Detector

### The Harris Corner Detection Algorithm (Szeliski)
1. Compute the horizontal and vertical derivatives of image Ix and Iy by convolving the original image with derivatives of Gaussian <br>
2. Compute three images corresponding to the outer products of these gradients<br>
3. Convolve each image with a larger Gaussian<br>
4. Compute a scalar interest measure R (corner response)<br>
5. Find local maxima above a certain threshold and report them as detected feature point locations

```python
from skimage.io import imread        # Import images
from skimage.color import rgb2gray   # Converts colors to gray
from scipy import signal as sig      # Spatial Derivativess
import numpy as np                   # Data analysis
import matplotlib.pyplot as plt      # Plot images
```

```python
# Convert to grayscale
img = imread('data/checkerboard.jpg')

imggray = rgb2gray(img)

plt.imshow(img)
plt.show()
```

1. Compute the horizontal and vertical gradients (Krishnan uses Sobel filtering)

```python
def gradient_x(imggray):
    ##Sobel operator kernels.
    kernel_x = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]])
    return sig.convolve2d(imggray, kernel_x, mode='same')
def gradient_y(imggray):
    kernel_y = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
    return sig.convolve2d(imggray, kernel_y, mode='same')

I_x = gradient_x(imggray)
I_y = gradient_y(imggray)
```

2. Compute three images corresp. to the outer products

```python
Ixx = I_x**2
Ixy = I_y*I_x
Iyy = I_y**2
```

3. Convolve each image with a larger gaussian <br>
4. Compute R

```python
height, width = imggray.shape
harris_response = []
window_size = 6
offset = window_size//2

k = 0.23                # Controls the threshold of corner response R
```

```python
# loop for all (x,y)
for y in range(offset, height-offset):
    for x in range(offset, width-offset):
        Sxx = np.sum(Ixx[y-offset:y+1+offset, x-offset:x+1+offset])
        Syy = np.sum(Iyy[y-offset:y+1+offset, x-offset:x+1+offset])
        Sxy = np.sum(Ixy[y-offset:y+1+offset, x-offset:x+1+offset])
        
        #Find determinant and trace, use to get corner response
        det = (Sxx * Syy) - (Sxy**2)
        trace = Sxx + Syy
        r = det - k*(trace**2)
        
        harris_response.append([x,y, r])
```

```python
img_copy = np.copy(img)

for response in harris_response:
    x, y, r = response
    if r > 0:
        img_copy[y,x] = [255,0,0]

        
plt.imshow(img_copy)
plt.show()
```
